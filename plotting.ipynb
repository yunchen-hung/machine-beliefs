```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['See', 'Hear', 'Discriminate colors', 'Discriminate distances',
'Think unsupported objects will fall', 'Think hidden objects are still there',
'Discriminate quantities', 'Prefer helping to not helping', 'Read']
values = ["<=.5", ".5-1", "<=.5", "<=.5", "<=.5", ".5-1", "2-4", "2-4" ,"2-4"]
custom_labels = ["<=.5", ".5-1", "1-2", "2-4"]

x_pos = [custom_labels.index(v) for v in values]
y_pos = [categories.index(v) for v in categories]

plt.scatter(x_pos, y_pos, marker='x', color='green')

plt.xticks(range(len(custom_labels)), custom_labels)
plt.yticks(range(len(categories)), categories)

plt.xlabel("Age")
plt.ylabel("Cognitive Ability")
plt.gca().invert_yaxis()
plt.title('Gemma 2 outputs (LLM)')
plt.show()
```
