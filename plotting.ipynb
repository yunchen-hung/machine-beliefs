### Gemma 3 (VLM)

"content": [{"type": "text", "text": "You are a human answering questions for a psychology survey."}]

prompt = f'{questions_on_empiricism[i]} Pick from image 1, 2, 3. \
            You must reply with either 1, 2, or 3 and specify the age of the child.'

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['See', 'Hear', 'Discriminate colors', 'Discriminate distances',
'Think unsupported objects will fall', 'Think hidden objects are still there',
'Discriminate quantities', 'Prefer helping to not helping', 'Read']
values = ["4-7", "7-10", "7-10", "7-10", "4-7", "7-10", "7-10", "7-10" ,"7-10"]
values_img = ['toddler', 'toddler', 'toddler', 'toddler', 'toddler', 'toddler', 'toddler', 'toddler','newborn']
custom_labels = ["<=.5", ".5-1", "1-2", "2-4", "4-7", "7-10"]
img_labels = ['newborn', 'toddler', 'schoolage_child']

label_map = {
    "newborn": "<=.5",
    "toddler": "1-2",
    "schoolage_child": "4-7"
}

second_row = []
for c in custom_labels:
    matched = [k for k, v in label_map.items() if v == c]
    second_row.append(matched[0] if matched else "")
two_line_labels = [f"{c}\n{s}" for c, s in zip(custom_labels, second_row)]

positions = range(len(custom_labels))

x_pos = [custom_labels.index(v) for v in values]
x_image = [2* img_labels.index(v) for v in values_img]
y_pos = [categories.index(v) for v in categories]

plt.scatter(x_pos, y_pos, marker='x', color='green', label='Numerical Categorization')
plt.scatter(x_image, y_pos, marker='x', color='red', label='Image Categorization')

plt.xticks(positions, two_line_labels)
plt.yticks(range(len(categories)), categories)

plt.xlabel("Age")
plt.ylabel("Cognitive Ability")
plt.gca().invert_yaxis()
plt.title('Gemma 3 outputs (VLM)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```
